```python
import random as r
ProductList = {'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35, 'p7':50,
               'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70,
               'p14':45}
```

```python
result_list=set()


while (len(result_list)<1000):
  size=r.randint(2,len(ProductList)-1)
  temp_list=r.sample(list(ProductList.keys()),size)
  temp_sum=sum([ProductList[i] for i in temp_list])
  if temp_sum in range(290,311):
     result_list.add(tuple(temp_list))

for i in result_list:
  print(list(i))

```
